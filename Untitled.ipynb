```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
```

```python
#function to get a product title

def get_title(soup):
    try:
        title = soup.find('span', attrs={'id': 'productTitle'})
        title_value = title.text
        title_string = title_value.strip()
    
    except AttributeError:
        title_string = ""
                     
    return title_string


#function to get product price
def get_price(soup):
    try:
        price = soup.find('span', attrs={"class": "a-offscreen"}).text.strip()
    
    except AttributeError:
        price = ""   
    
    return price

#fucntion to get product rating

def get_rating(soup):
    try:
        rating = soup.find("span", attrs={"class": "a-icon-alt"}).text.strip()
    
    except AttributeError:
        rating = ""   
    
    return rating

#function to get number of reviews

def get_NumberOfReviews(soup):
    try:
        number_reviews = soup.find("span", attrs={'id': "acrCustomerReviewText"}).text.strip()
        
    except AttributeError:
        number_reviews = ""   
    
    return number_reviews


#function to get shipping and import fees

def get_shipping(soup):
    try:
        shipping = soup.find("span", attrs={"class": "a-size-base a-color-secondary"}).text.strip()
    
    except AttributeError:
        shipping = ""
        
    return shipping
        
        
```

```python
if __name__ == '__main__':
    
    HEADERS = ({
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36',
    'Accept-Language': 'en-US, en;q=0.5'})
     
        #Amazon URL
    URL =' https://www.amazon.com/s?k=playstation+5&ref=nb_sb_noss_1'
    
    #Soup containing all the data
    webpage = requests.get(URL, headers=HEADERS)
    
    # Soup Object containing all data
    soup = BeautifulSoup(webpage.content, "html.parser")
    
    #Fecthing all the playstaion 5 links
    links = soup.find_all("a", attrs={'class': 'a-link-normal s-no-outline'})
    
    #storing the links
    links_list = []
    
    #Loop for getting all the links
    for link in links:
        links_list.append(link.get('href'))
        
        
    dictio = {"price":[], "title":[], "rating": [], "reviews":[], "shipping":[] }
    
    
    #Looping to get product details from each link
    for link in links_list:
        try:
            
            new_webpage = requests.get("https://www.amazon.com" + link, headers=HEADERS)
            new_soup = BeautifulSoup(new_webpage.content, "html.parser")
            dictio['price'].append(get_price(new_soup))
            dictio['title'].append(get_title(new_soup))
            dictio['rating'].append(get_rating(new_soup))
            dictio['reviews'].append(get_NumberOfReviews(new_soup))
            dictio['shipping'].append(get_shipping(new_soup))
        except (requests.exceptions.RequestException, ConnectionError) as e:
            continue
        
        
        
          
    
```

```python
df = pd.DataFrame(dictio)
df.to_csv('data.csv', index=False)

```
